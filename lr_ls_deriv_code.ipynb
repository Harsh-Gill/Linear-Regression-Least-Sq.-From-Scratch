```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
```

# Random Points generated 

```python
# random points of scatter

# point list

pts = np.array([[1,1],[2,3],[3,4],[4,7 ],[5,6],[6,8],[7,5]])

import matplotlib.pyplot as plt
f, ax = plt.subplots(1)

ax.scatter(pts[:,0],pts[:,1])
ax.set_ylim(bottom=0)
plt.show(f)
```

# Defining Linear Regression Model from Derivation 


```python
## Define Linear Regression Model

# defining scatter points
X = pts[:,0] 
Y = pts[:,1] 

# mean of our inputs and outputs
x_mean = np.mean(X)
y_mean = np.mean(Y)

#total number of values
n = len(X)

sum_yi_xi_n = 0
sum_xi = 0
sum_yi = 0
sum_xisq_n = 0

for i in range(n):
    sum_yi_xi_n += n*Y[i]*X[i]
    sum_xi += X[i]
    sum_yi += Y[i]
    
    sum_xisq_n += n*(X[i])**2

b1 = (sum_yi_xi_n - sum_xi*sum_yi) / (sum_xisq_n - sum_xi**2)


b0 = sum_yi/n - b1*(sum_xi)/n
#printing the b1 and b0

print(b1, b0)


```

## Visualizing Best Fit Line 

```python
#plotting values 
x_max = np.max(X)
x_min = np.min(X)
#calculating line values of x and y
x = np.linspace(x_min, x_max, 10)
y = b0 + b1 * x
#plotting line 
plt.plot(x, y, color='#00ff00', label='Linear Regression')
#plot the data point
plt.scatter(X, Y, color='#ff0000', label='Data Point')
# x-axis label


plt.show()
```

# RMSE Error

```python
# Using Root Mean Squared Error
rmse = 0
for i in range(n):
    y_pred=  b0 + b1* X[i]
    rmse += (Y[i] - y_pred) ** 2
    
rmse = np.sqrt(rmse/n)
print(rmse)
```
